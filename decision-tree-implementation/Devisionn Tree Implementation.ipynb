```python
import pandas as pd         # To work easily with tabular data (dataframes)
import numpy as np          # For handling numerical data and arrays
import matplotlib.pyplot as plt  # For making visual plots of our data

%matplotlib inline
# Ensures plots show up inside the notebook
```

```python
from sklearn.datasets import load_iris   # Access the classic Iris dataset for demo and practice

iris = load_iris() # Load these flower measurements and species into our workspace
```

```python
import seaborn as sns # For advanced data visualization and quick access to sample datasets
df = sns.load_dataset('iris') # Load seaborn's version of the Iris dataset into a DataFrame for easy use
df.head() # Preview the first rows to understand its structure
```

```python
#Independent features (i/p) and dependent features (o/p)

X = df.iloc[:,:-1] 
y = iris.target # Target labels: which species (scikit-learn's format)
```

```python
X, y
```

```python
# Train and test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42) # Split data: 67% train, 33% test
```

```python
X_train # Inspect the training features
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Post Prunning (unrestricted)

treemodel = DecisionTreeClassifier()
treemodel.fit(X_train, y_train) # Fit the model to our training data
```

```python
from sklearn import tree
plt.figure(figsize=(15,10)) # Set size for the plot
tree.plot_tree(treemodel,filled=True) # Visualize the trained decision tree
```

```python
treemodel1 = DecisionTreeClassifier(max_depth=3) # to avoid overfitting
treemodel1.fit(X_train, y_train) # Fit pruned tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel1,filled=True)
```

```python
# Prediction
y_pred = treemodel.predict(X_test)
y_pred # Show predicted labels
```

```python
from sklearn.metrics import accuracy_score, classification_report
score = accuracy_score(y_pred,y_test) 
print(score)
print(classification_report(y_pred,y_test))
```

```python
# Define a grid for hyperparameter tuning (to find the best model)

parameter = {
 'criterion': ['gini', 'entropy', 'log_loss'],     # how the tree splits (Gini Index, Entropy, Log Loss)
 'splitter': ['best', 'random'],                  # how to choose the split at each node
 'max_depth': [1, 2, 3, 4, 5],                    # how deep the tree can grow
 'max_features': ['sqrt', 'log2']                # features considered at each split (fixed below 👇)
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
treemodel = DecisionTreeClassifier()
cv = GridSearchCV(treemodel, param_grid=parameter, cv=5, scoring='accuracy') # Set up grid search across all param combos
cv.fit(X_train, y_train)                   # Run the search on training data
print("Best Parameters:", cv.best_params_) # Show the optimal parameter values
print("Best Score (Training Accuracy):", cv.best_score_)
best_model = cv.best_estimator_            # Retrieve the best model found by GridSearch
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Make predictions with the best-found tree
y_pred = best_model.predict(X_test)

# Evaluate the tuned model
print("Test Accuracy:", accuracy_score(y_test, y_pred))            # Final model accuracy on new data
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))     # Show which classes get mixed up
print("Classification Report:\n", classification_report(y_test, y_pred))   # Detailed stats
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))                                       # Make tree visualization large and clear
plot_tree(best_model, 
          feature_names=iris.feature_names, 
          class_names=iris.target_names, 
          filled=True, 
          rounded=True)                                           # Show how the final model makes decisions
plt.title("Decision Tree Visualization")
plt.show()
```
